```python
from bitcoin_lib import *
```

### Data ingestion

```python
df = load_bitcoin_data()
```

# EDA

### Simple Moving Average 

```python
columns = ['close', 'sma15_close', 'sma45_close', 'sma90_close', 'sma200_close']
plot(df=df, columns=columns, xlabel="Date", ylabel="Closing Price", title="Simple Moving Average for Close")
```

### Exponential Moving Average

```python
columns = ['close', 'ema15_close', 'ema45_close', 'ema90_close', 'ema200_close']
plot(df=df, columns=columns, xlabel="Date", ylabel="Closing Price", title="Exponential Moving Average for Close")
```

### RSI 

```python
columns = ['rsi_15d', 'rsi_45d', 'rsi_90d', 'rsi_200d']
plot(df=df, columns=columns, xlabel="Date", ylabel="Closing Price", title="RSI for Close")
```
